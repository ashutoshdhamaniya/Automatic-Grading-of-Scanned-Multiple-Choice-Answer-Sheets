```python
import cv2
import numpy as np 
import imutils
from imutils import contours
import import_ipynb
import four_point
```

```python
Answer_key= {0:1, 1:4, 2:0, 3:2, 4:3}
```

```python
# read image
image= cv2.imread("test_images/test_01.jpeg")
cv2.imshow("omr", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
# cv2.imshow("omr", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# convert image to gray scale image
gray_scale_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# cv2.imshow("omr", gray_scale_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# apply filter to remove noise
blurred_image= cv2.GaussianBlur(gray_scale_image, (5,5), 0)

cv2.imshow("omr", blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# canny edge detection to detect the edges in the given image
edges_in_image = cv2.Canny(blurred_image, 75, 200)

cv2.imshow("omr", edges_in_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# to extract the rectangle document from the image 
contours_in_image = cv2.findContours(edges_in_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours_in_image= imutils.grab_contours(contours_in_image)
```

```python
# cv2.drawContours(image, contours_in_image, -1, (0,255,0), 3)

# cv2.imshow("Contours Images", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# omr sheet countours initially None
omr_contours = None

# to check that if countours find in the image or not
if len(contours_in_image)>0: 
    # sort countours in descending order
    contours_in_image= sorted(contours_in_image, key=cv2.contourArea, reverse=True)
    
    for contour in contours_in_image:
        rectangle_perimeter = cv2.arcLength(contour, True)
        rectangle_perimeter_approx = cv2.approxPolyDP(contour, 0.02*rectangle_perimeter, True)

        # shape is quadrilateral so compare with 4
        if(len(rectangle_perimeter_approx) == 4):
            omr_contours= rectangle_perimeter_approx
            break
```

```python
# cv2.drawContours(image, omr_contours, -1, (0,255,0), 3)

# cv2.imshow("Contours Images", image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# extracting the OMR sheet and removing all the unnecessary part
extract_omr=  four_point.four_point_transform(gray_scale_image, omr_contours.reshape(4, 2))

cv2.imshow("Extracted OMR", extract_omr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
binary_image= cv2.threshold(extract_omr, 0, 255, cv2.THRESH_BINARY_INV|cv2.THRESH_OTSU)[1]

cv2.imshow("Binary Image", binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# again find contours to extract the round part in the OMR sheet

contours_in_OMR_sheet= cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours_in_OMR_sheet= imutils.grab_contours(contours_in_OMR_sheet)

options_contours= []
```

```python
for contour in contours_in_OMR_sheet:
    (x, y, width, height)= cv2.boundingRect(contour)
    width_height_ratio= width/ float(height)
    
    print(width_height_ratio)
    
    # checking the width, height and width_height_ration to select the contours of options
    if width>=20 and height>=20 and width_height_ratio >=0.9 and width_height_ratio<= 1.1:
        options_contours.append(contour)
     
    
# cv2.drawContours(binary_image, options_contours, -1, (255,255,255), 3)

# cv2.imshow("Options Contours", binary_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
# enumerate(start_index -> end_index, step_size(number_of_options)
correct_answers= 0
contour_len = len(options_contours)

print(contour_len)

for (q, i) in enumerate(np.arange(contour_len, 0, -5)):
    # sort contours from left to right
    sorted_options_contours= contours.sort_contours(options_contours[i-5:i]) [0]
    marked_answer= None
    
    cv2.drawContours(binary_image, sorted_options_contours, -1, (255,255,255), 3)

    cv2.imshow("Options Contours", binary_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    for (j,c) in enumerate(sorted_options_contours):
        mask = np.zeros(binary_image.shape, dtype="uint8")
#         print(mask)
        cv2.drawContours(mask, [c], -1, 255, -1)
        
        cv2.imshow("j", mask)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        
        mask= cv2.bitwise_and(binary_image, binary_image, mask=mask)
        total= cv2.countNonZero(mask)
        
        cv2.imshow("j", mask)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        if marked_answer is None or total >marked_answer[0]:
            marked_answer= (total, j)
            k= Answer_key[q]    
            print(marked_answer, k)
            if k == marked_answer[1]:
                correct_answers= correct_answers+1
                
        
            
score = (correct_answers/5.0)*100
print(correct_answers)
print(score)
```
